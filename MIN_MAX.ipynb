```python
import math
board=[" "," "," "," "," "," "," "," "," "," "]
def print_board(board):
    row1=board[0:3]
    row2=board[3:6]
    row3=board[6:9]
    print("|".join(row1))
    print("|".join(row2))
    print("|".join(row3))
       
def available_moves(b):
    return [i for i,spot in enumerate(b) if spot==" "]
def winner(b):
    win_combos=[
        (0,1,2),(3,4,5),(6,7,8),
        (0,3,6),(1,4,7),(2,5,8),
        (0,4,8),(2,4,6)
    ]
    for (x,y,z) in win_combos:
        if b[x]==b[y]==b[z] and b[x]!=" ":
            return b[x]
    return None
def terminal_state(b):
    return winner(b) or " " not in b
def utility(b):
    win=winner(b)
    if win=="X":
        return 10
    elif win=="O":
        return -10
    return 0
def minmax(b,depth,is_max):
    if terminal_state(b):
        return utility(b)
    if is_max:
        best=-math.inf
        for move in available_moves(b):
            b[move]="X"
            score=minmax(b,depth+1,False)
            b[move]=" "
            best=max(best,score)
        return best
    else:
        best=math.inf
        for move in available_moves(b):
            b[move]="O"
            score=minmax(b,depth+1,True)
            b[move]=" "
            best=min(best,score)
        return best
def best_move():
    best_val=-math.inf
    move=None
    for i in available_moves(board):
        board[i]="X"
        score=minmax(board,0,False)
        board[i]=" "
        if score>best_val:
            best_val=score
            move=i
    return move
def play():
    print("Tic Tac Toe:nee(0) vs AI(x)")
    #print_board()
   
    while not terminal_state(board):
        human=int(input("enter your move (0-8):"))
        if board[human]!=" ":
            print("Invalid move,try again")
            continue
        board[human]="O"
        if terminal_state(board):
            break
        ai=best_move()
        board[ai]="X"
        print(f"AI choose position {ai}")
        print_board(board)
    print("final game position:")
    print_board(board)
    win=winner(board)
    if win=="X":
        print('AI win!')
    elif win=="O":
        print("naama jeichiutom maara!")
    else:
        print("sirappaana aal nee iruvarum samam!")
play()   
```
